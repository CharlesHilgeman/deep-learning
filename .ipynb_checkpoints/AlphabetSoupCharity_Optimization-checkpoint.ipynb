```python
import os
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
import pandas as pd
import tensorflow as tf
```

```python
alphabet_df = pd.read_csv('Resources/charity_data.csv')
alphabet_df.head()
```

```python
# replace ask_amount rows >5000
for row in alphabet_df['ASK_AMT']:
    if row > 5000:
        alphabet_df['ASK_AMT'] = alphabet_df['ASK_AMT'].replace(row,">5000")
    else:
        alphabet_df['ASK_AMT'] = alphabet_df['ASK_AMT'].replace(row,"<=5000")
```

```python
alphabet_types_to_replace = ['T17','T15','T29','T14','T25','T2','T12','T13','T9']

# Replace in dataframe
for app in alphabet_types_to_replace:
    alphabet_df['APPLICATION_TYPE'] = alphabet_df['APPLICATION_TYPE'].replace(app,"Other")

# Check to make sure binning was successful
alphabet_df['APPLICATION_TYPE'].value_counts()
```

```python
class_edit = pd.DataFrame(alphabet_df['CLASSIFICATION'].value_counts())
```

```python
class_edit = class_edit.reset_index()
class_edit = class_edit.loc[class_edit['CLASSIFICATION'] < 1500]
classifications_to_replace = class_edit['index'].tolist()

# Replace in dataframe
for cls in classifications_to_replace:
    alphabet_df['CLASSIFICATION'] = alphabet_df['CLASSIFICATION'].replace(cls,"Other")
    
# Check to make sure binning was successful
alphabet_df['CLASSIFICATION'].value_counts()
```

```python
alphabet_df = alphabet_df.drop(columns = ['EIN', 'NAME','STATUS','SPECIAL_CONSIDERATIONS'])
alphabet_df.head()
```

```python
alphabet_cat = alphabet_df.dtypes[alphabet_df.dtypes == "object"].index.tolist()
alphabet_cat
```

```python
from sklearn.preprocessing import OneHotEncoder
# Convert categorical data to numeric with `pd.get_dummies`

enc = OneHotEncoder(sparse=False)
encode_df = pd.DataFrame(enc.fit_transform(alphabet_df[alphabet_cat]))
encode_df.columns = enc.get_feature_names(alphabet_cat)
encode_df.head()
```

```python
alphabet_df = alphabet_df.merge(encode_df,left_index=True, right_index=True)
alphabet_df = alphabet_df.drop(alphabet_cat,1)
alphabet_df.head()
```

```python
y = alphabet_df['IS_SUCCESSFUL'].values
X = alphabet_df.drop(columns = 'IS_SUCCESSFUL').values

X_train, X_test, y_train, y_test = train_test_split(X, y, random_state = 42)
```

```python
scaler = StandardScaler().fit(X_train)

X_train_scaled = scaler.transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

```python
# Define the model - deep neural net, i.e., the number of input features and hidden nodes for each layer.
nn_model = tf.keras.models.Sequential()

# First hidden layer
nn_model.add(tf.keras.layers.Dense(units=80, activation = 'relu', input_dim=41))

# Second hidden layer
nn_model.add(tf.keras.layers.Dense(units=30, activation = 'tanh'))

# Output layer
nn_model.add(tf.keras.layers.Dense(units=1, activation = 'sigmoid'))

# Check the structure of the model
nn_model.summary()
```

```python
nn_model.compile(loss = 'binary_crossentropy', optimizer = 'adam', metrics = ['accuracy'])
```

```python
fit_model = nn_model.fit(X_train_scaled, y_train, epochs = 200)
fit_model_df = pd.DataFrame(fit_model.history)
```

```python
fit_model_df = fit_model_df.iloc[::5, :]
fit_model_df['accuracy'].max()
```

I tried multiple different techniques/updates but could not get my accuracy score above 75%. The closest I could get was just above 74%

```python
model_loss, model_accuracy = nn_model.evaluate(X_test_scaled,y_test,verbose=2)
print(f"Loss: {model_loss:.3f}, Accuracy: {model_accuracy:.3f}")
```

```python
nn_model.save_weights('output/optimized_nn_model_weights.h5')
```
